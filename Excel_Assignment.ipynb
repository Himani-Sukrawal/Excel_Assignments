**Question 1: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly
explain with examples.**

**Answer :** Here’s a clear difference between Paste and Paste Special in Excel -

**1. Paste (Normal Paste)**

Definition: Copies everything (values, formulas, formatting, comments, etc.) from the copied cell(s) to the destination.

Example:

If cell A1 = 5+5 (formula), and you copy and paste it into B1, then B1 will also show 10 but still contain the formula =5+5.

Formatting like bold, font color, and borders will also be carried over.

**2. Paste Special**

Definition: Gives more control over what to paste (only values, only formulas, only formatting, etc.).

Examples:

Paste Special → Values:
If A1 = 5+5, pasting as Values into B1 will only paste 10, not the formula.

Paste Special → Formats:
Copies only the formatting (font, color, border) without the numbers or formulas.

Paste Special → Multiply / Add / Subtract / Divide:
You can paste values with an operation. For example: If A1 = 10 and you copy 2, then use Paste Special → Multiply on A1, the result will become 20.

**In short:**

Paste = everything.

Paste Special = only specific elements (values, formulas, formatting, or operations).

**Question 2: Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes'
in Excel.**

**Answer :** Please refer to below explanation:

**Freeze Panes**

Function:

Keeps specific rows or columns visible while you scroll through the rest of the worksheet.

Usefulness:

Helpful when working with large datasets where you always want to see headers or key labels.

Examples:

Freeze Top Row → Keeps the first row (usually column headings) fixed while scrolling down.

Freeze First Column → Keeps the first column (e.g., employee IDs) visible while scrolling sideways.

Freeze Panes (custom) → Freezes rows above and columns to the left of the active cell.

**Split Panes**

Function:

Divides the worksheet window into two or four separate sections, each with its own scroll bar.

Usefulness:

Allows you to view and compare different parts of the same worksheet at the same time.

Examples:

Splitting a sheet into 2 horizontal sections → you can see the top headers and bottom data simultaneously.

Splitting into 4 quadrants → helpful for comparing values in widely separated rows and columns without losing track.

**In short:**

Freeze Panes = Keeps headers/labels visible while scrolling.

Split Panes = Creates multiple scrollable sections of the same sheet for side-by-side viewing.

**Question 3: Explain the difference between inserting a new row and inserting a new
column in Excel. Can you insert multiple rows or columns at once?**

**Answer:** Here’s the clear explanation:

**1. Inserting a New Row**

What it does: Adds a new horizontal row above the selected row.

Effect: Shifts existing rows down.

Example: If you select Row 5 and insert a new row, the new row will become Row 5, and the old Row 5 (and below) will move down by one.

**2. Inserting a New Column**

What it does: Adds a new vertical column to the left of the selected column.

Effect: Shifts existing columns right.

Example: If you select Column C and insert a new column, the new column will become Column C, and the old Column C (and rightward) will shift to the right.

**3. Can You Insert Multiple Rows or Columns at Once?**

Yes, multiple rows or columns can be inserted at once.

Rows: Select the number of existing rows equal to how many you want to add, then right-click → Insert.

Example: Select 3 rows, right-click, and choose Insert → Excel will add 3 new rows above the first selected row.

Columns: Same method—select multiple columns, right-click → Insert, and Excel will insert that many columns to the left of the first selected column.

**In short:**

Row insertion = adds space horizontally (shifts rows down).

Column insertion = adds space vertically (shifts columns right).

You can insert multiple rows/columns at once by selecting them before inserting.

**Question 4: What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.**

**Answer:** Logical functions in Excel are functions that test conditions and return a logical value — either TRUE or FALSE. They are widely used in decision-making, data analysis, and conditional calculations.

**Examples of Logical Functions:**

**1. IF Function**

Syntax: =IF(condition, value_if_true, value_if_false)

Use: Checks a condition and returns one value if it’s true, another if it’s false.

Example:
=IF(A1>=40, "Pass", "Fail")
→ If the value in cell A1 is 40 or above, it shows Pass, otherwise Fail.

**2. AND Function**

Syntax: =AND(condition1, condition2, …)

Use: Returns TRUE only if all conditions are true.

Example:
=AND(A1>=40, B1>=40)
→ Returns TRUE only if both A1 and B1 are greater than or equal to 40.

**3. OR Function (another common one)**

Syntax: =OR(condition1, condition2, …)

Use: Returns TRUE if any one of the conditions is true.

Example:
=OR(A1>=60, B1>=60)
→ Returns TRUE if either A1 or B1 is greater than or equal to 60.

**In short:**
Logical functions like IF, AND, OR help test conditions and make decisions in Excel. They are useful for grading, filtering, validations, and conditional analysis.


**Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function.**

**Answer:** Purpose of XLOOKUP:

The XLOOKUP function in Excel is used to search for a value in a range or array and return a corresponding value from another range.

It is more flexible and powerful than the older VLOOKUP or HLOOKUP functions.

Syntax:
=XLOOKUP(lookup_value, lookup_array, return_array, [if_not_found], [match_mode], [search_mode])

**How XLOOKUP Differs from VLOOKUP:**

**1. Search Direction:**

VLOOKUP only searches vertically (in columns).

XLOOKUP can search both vertically and horizontally.

**2. Lookup Position:**

VLOOKUP requires the lookup column to be the first column in the range.

XLOOKUP works anywhere — the lookup column does not need to be the first.

**3.Exact Match vs Approximate Match:**

VLOOKUP defaults to approximate match (if you forget to add “FALSE”).

XLOOKUP defaults to exact match, reducing errors.

**4.Return Multiple Values:**

XLOOKUP can return multiple columns (spill results).

VLOOKUP can only return one column at a time.

**5.Error Handling:**

XLOOKUP has an optional [if_not_found] argument to display custom messages like “Not Found.”

VLOOKUP shows #N/A if a value is missing.

**Example:**

VLOOKUP:

=VLOOKUP(101, A2:C10, 2, FALSE) → Looks for employee ID 101 in column A and returns value from column B.

XLOOKUP:

=XLOOKUP(101, A2:A10, B2:B10, "Not Found") → Looks for employee ID 101 in column A and returns corresponding value from column B (or shows Not Found if missing).

**In short:**

VLOOKUP = older, limited (only vertical, needs first column).

XLOOKUP = modern, flexible, works in both directions, defaults to exact match, and supports error handling.

Steps:

Select your data range.

Go to Home → Find & Select → Go To Special → Blanks → OK.

Excel will highlight all empty cells in the selected range.

**b) ISBLANK Function**

Purpose: Check if a specific cell is blank. Returns TRUE if blank, FALSE otherwise.

Formula Example:

=ISBLANK(B2)


Use Case: Create a new column that flags missing data for analysis.
Example: You can filter all rows where ISBLANK is TRUE to review missing entries.

**c) COUNTBLANK Function**

Purpose: Count the number of blank cells in a column or range.

Formula Example:

=COUNTBLANK(B2:B100)


Use Case: Gives a quick summary of how many values are missing in a column. Useful for reporting or deciding whether to delete or impute missing values.

**2. Handling Missing Data**

**a) Numerical Columns (e.g., Age, Salary)**

Replace with average/median:

Delete rows: Only if missing data is minimal and imputation is not suitable.
